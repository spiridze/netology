7 Проверить гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом


```python
import scipy.stats as st
```

```python
high_salary_employees = hr[hr['salary'] == 'high']['average_montly_hours']
low_salary_employees = hr[hr['salary'] == 'low']['average_montly_hours']
```

```python
# нулевая гипотеза: сотрудники с высоким окладом проводят на работе в среднем столько же времени, как и сотрудники с низким окладом

stat, p = st.ttest_ind(high_salary_employees, low_salary_employees)

print(stat, p)

if p < 0.05:
    print("Сотрудники с высоким окладом проводят на работе в среднем больше или меньше времени по сравнению с сотрудниками с низким окладом")
else:
    print("Сотрудники с высоким окладом проводят на работе в среднем столько же времени, как и сотрудники с низким окладом")
```

```python
print(high_salary_employees.mean())
print(low_salary_employees.mean())
```

8 Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):

● Доля сотрудников с повышением за последние 5 лет\
● Средняя степень удовлетворенности\
● Среднее количество проектов


```python
left_promotion_rate = (hr[hr['left'] == 1]['promotion_last_5years'].sum() / hr[hr['left'] == 1]['promotion_last_5years'].count()) * 100
stayed_promotion_rate = (hr[hr['left'] == 0]['promotion_last_5years'].sum() / hr[hr['left'] == 0]['promotion_last_5years'].count()) * 100

print("Доля уволившихся с повышением за последние 5 лет (%):", left_promotion_rate)
print("Доля не уволившихся с повышением за последние 5 лет (%):", stayed_promotion_rate)
```

```python
grouped_by_left_employees = hr.groupby('left')

grouped_by_left_employees
```

```python
satisfaction = grouped_by_left_employees['satisfaction_level'].mean()

print(f'Средняя степень удовлетворенности среди оставшихся (0) и уволившихся (1) сотрудников: {satisfaction}')
```

```python
projects = grouped_by_left_employees['number_project'].mean()

print(f'Среднее количество проектов среди оставшихся (0) и уволившихся (1) сотрудников: {projects}')
```

9 Разделить данные на тестовую и обучающую выборки\
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)\
Оценить качество модели на тестовой выборки

```python
X = hr[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours',
       'time_spend_company', 'Work_accident', 'promotion_last_5years']]
y = hr['left']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, lda.predict(X_test))

print(f'Точность модели {accuracy*100:.2f}%')
```
